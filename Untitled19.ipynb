Q6. Train your model on different train test ratio and train the models and note down their accuracies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

data = data[['FUELCONSUMPTION_COMB', 'CO2EMISSIONS']]

data = data.dropna()

X = data[['FUELCONSUMPTION_COMB']]
y = data['CO2EMISSIONS']


ratios = [0.2, 0.3, 0.4]


results = {}

for ratio in ratios:

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ratio, random_state=42)

    model = LinearRegression()

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    results[ratio] = mse
    print(f"Test size ratio {ratio*100}% - Mean Squared Error: {mse}")

    plt.figure()
    plt.scatter(X_test, y_test, color='blue', label='Actual data')
    plt.plot(X_test, y_pred, color='red', label='Predicted line')
    plt.xlabel('Fuel Consumption (Combined)')
    plt.ylabel('CO2 Emissions')
    plt.title(f'Fuel Consumption vs CO2 Emissions (Test size ratio {ratio*100}%)')
    plt.legend()
    plt.show()

print("\nMean Squared Errors for different train-test ratios:")
for ratio, mse in results.items():
    print(f"Test size ratio {ratio*100}%: Mean Squared Error = {mse}")
```
